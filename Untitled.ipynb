```python
from config import zws_id
import requests
from xml.etree import ElementTree as ET
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
target_url = f"http://www.zillow.com/webservice/GetRegionChildren.htm?zws-id={zws_id}&state=ca&city=sandiego&childtype=neighborhood"
```

```python
response = requests.get(target_url)
```

```python
root = ET.fromstring(response.content)
```

```python
count = int(root[2][2][0].text)
id = []
name = []
zindex = []
url = []
lati = []
long = []

import pandas as pd
df = pd.DataFrame({"id":id,
                   "name":name,
                   "price":zindex,
                   "url":url,
                   "lati":lati,
                   "long":long})

row_list = []

for i in range(count):
    dict_0 = {}
    try:
        for j in range(6):
            #print(f"#{j}:",end='')
            #print(f"{root[2][2][1+i][j].tag}:",end="") 
            #print(root[2][2][1+i][j].text)
            dict_0.update({root[2][2][1+i][j].tag:root[2][2][1+i][j].text})
    except IndexError:
        print("missing info")
    row_list.append(dict_0)

df = pd.DataFrame(row_list)

print("DataFrame has been generated!")
```

```python
df = df.dropna(how='any')
```

```python
import gmaps
gmaps.configure(api_key=gkey)   
```

```python
coordinate = []
region = []
zvalue = []
for index, row in df.iterrows():
    lat = row["latitude"]
    lon = row["longitude"]
    name = row["name"]
    zindex = row["zindex"]
    coordinate.append((lat,lon))
    region.append(name)
    zvalue.append(zindex)
    
    
```

```python

fig = gmaps.figure()
```

```python
locations = df[["latitude", "longitude"]].astype(float)
heat_layer = gmaps.heatmap_layer(locations, weights=zvalue, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 1)

heat_layer.dissipating = False
heat_layer.max_intensity = max(zvalue)
heat_layer.point_radius = 0.02


fig.add_layer(heat_layer)

fig
```

```python
markers = gmaps.marker_layer(coordinate,info_box_content=[f"Zillow Estimated Value: {zv}" for zv in zvalue])
# Add the layer to the map
fig.add_layer(markers)
fig
```

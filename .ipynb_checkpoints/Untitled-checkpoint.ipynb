```python
from config import zws_id
import requests
from xml.etree import ElementTree as ET
target_url = f"http://www.zillow.com/webservice/GetRegionChildren.htm?zws-id={zws_id}&state=ca&city=sandiego&childtype=neighborhood"
```

```python
target_url_2 = f"http://www.zillow.com/webservice/GetRegionChildren.htm?zws-id={zws_id}&state=ca&city=sandiego&childtype=city"
```

```python
response = requests.get(target_url)
```

```python
root = ET.fromstring(response.content)
```

```python
count = int(root[2][2][0].text)
id = []
name = []
zindex = []
url = []
lati = []
long = []

import pandas as pd
df = pd.DataFrame({"id":id,
                   "name":name,
                   "price":zindex,
                   "url":url,
                   "lati":lati,
                   "long":long})

row_list = []

for i in range(count):
    dict_0 = {}
    try:
        for j in range(6):
            #print(f"#{j}:",end='')
            #print(f"{root[2][2][1+i][j].tag}:",end="") 
            #print(root[2][2][1+i][j].text)
            dict_0.update({root[2][2][1+i][j].tag:root[2][2][1+i][j].text})
    except IndexError:
        print("missing info")
    row_list.append(dict_0)

df = pd.DataFrame(row_list)

print("DataFrame has been generated!")
```

```python
df = df.dropna(how='any')
```

```python
plt.figure()
plt.plot(x,sum)
```

```python
def f_1(x, A, B):
    return A*x + B

x1 = np.arange(2010, 2017, 0.1)
A1, B1 = optimize.curve_fit(f_1, x, sum)[0]
y1 = A1*x1 + B1
plt.plot(x1,y1,"k.")
```
